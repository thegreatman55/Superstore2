```python
def classify_columns(df1):
    numerical_columns = df1.select_dtypes(include=['number']).columns.tolist()
    categorical_columns = df1.select_dtypes(exclude=['number']).columns.tolist()
    
    return numerical_columns, categorical_columns
```

```python
numerical_columns, categorical_columns = classify_columns(df1)

print("Numerical Columns:", numerical_columns)
print("Categorical Columns:", categorical_columns)
```

```python
excluded_columns = {"Order ID", "Product Name", "Product ID", "Order Date", "Ship Date", "Customer ID", "Customer Name"}
categorical_columns = [col for col in categorical_columns if col not in excluded_columns]
```

```python
unique_values = {col: df1[col].unique().tolist() for col in categorical_columns}
```

##### Profitability by City

```python
profit_by_city = df1.groupby("City")["Profit"].sum().reset_index()
highest_profit_cities = profit_by_city.sort_values(by="Profit", ascending=False).head(10)
sales_by_city = df1.groupby("City")["Sales"].sum().reset_index()
city_sales_profit = pd.merge(sales_by_city, profit_by_city, on="City")
cities_high_sales_low_profit = city_sales_profit[(city_sales_profit["Sales"] > city_sales_profit["Sales"].median()) & (city_sales_profit["Profit"] < city_sales_profit["Profit"].median())]
```

##### Visualization for most profitable cities

```python
plt.figure(figsize=(14, 6))
sns.barplot(x="City", y="Profit", data=highest_profit_cities, palette="viridis")
plt.xlabel("City")
plt.ylabel("Total Profit")
plt.title("Top 10 Most Profitable Cities")
plt.xticks(rotation=90)
plt.show()
```

##### Visualization for Cities with High Sales but low profits

```python
plt.figure(figsize=(14, 6))
sns.barplot(x="City", y="Sales", data=cities_high_sales_low_profit, palette="coolwarm")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.title("Cities with High Sales but Low Profit")
plt.xticks(rotation=90)
plt.show()
```

##### Regional Discounts*: Are there regions where higher discounts are offered more frequently? How do these discounts affect sales and profit in those regions?

```python
discount_by_region = df1.groupby("Region")["Discount"].mean().reset_index()
sales_profit_discount_region = df1.groupby("Region").agg({"Discount": "mean", "Sales": "sum", "Profit": "sum"}).reset_index()
```

##### Visualization of Discount by Region

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Region", y="Discount", data=discount_by_region, palette="coolwarm")
plt.xlabel("Region")
plt.ylabel("Average Discount")
plt.title("Average Discount Offered by Region")
plt.show()
```

##### Visualization for Sales and Profit by Discounted Regions

```python
fig, ax1 = plt.subplots(figsize=(12, 6))
ax2 = ax1.twinx()
sns.barplot(x="Region", y="Sales", data=sales_profit_discount_region, ax=ax1, color="blue", alpha=0.6)
sns.lineplot(x="Region", y="Profit", data=sales_profit_discount_region, ax=ax2, color="red", marker="o")
ax1.set_xlabel("Region")
ax1.set_ylabel("Total Sales", color="blue")
ax2.set_ylabel("Total Profit", color="red")
plt.title("Sales and Profit Across Regions with Discount Levels")
plt.show()
```
